```python
import numpy as np
import pandas as pd
```

```python
distance_path = "Embeddings/pairwise_distances.csv"
cluster_path = "ResNetClusteringResults/weighted-cosine_10_clusters.csv"
result_path = "cluster_representations.csv"
```

```python
distance_df = pd.read_csv(distance_path,
                          index_col="anime_id")
cluster_df = pd.read_csv(cluster_path,
                         index_col="anime_id")
print(distance_df.shape,cluster_df.shape)
print(distance_df.head())
print(distance_df.values[:,0])
```

```python
cluster = cluster_df["cluster"].values
max_cluster = cluster.max()
print(distance_df.shape,)
l=[]
for i in range(1,max_cluster+1):
    cluster_observations = np.arange(distance_df.shape[0])[cluster_df["cluster"] ==i]
    cluster_distances = distance_df.iloc[cluster_observations,cluster_observations]
    cluster_sum_distances = cluster_distances.values.sum(axis=1)
    min=cluster_sum_distances.argmin()
    l.append(cluster_distances.index[min])
```

```python
import kagglehub
svanoo_myanimelist_dataset_path = kagglehub.dataset_download('svanoo/myanimelist-dataset')
animes_path = f'{svanoo_myanimelist_dataset_path}/anime.csv'
anime_df=pd.read_csv(animes_path,sep="\t")
anime_df.set_index('anime_id',inplace=True)
anime_df = anime_df.loc[ l ]
```

```python
anime_df.to_csv(result_path)
```
