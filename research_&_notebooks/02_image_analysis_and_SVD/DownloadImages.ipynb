```python
anime.sort_values("total_count",
                  inplace=True,ascending=False)
print(anime.head())
anime_to_consider = anime.head(2000).copy()
anime_to_consider.set_index('anime_id', inplace=True)
```

```python
import io

images_folder = "Anime_images"
from PIL import Image
os.makedirs(images_folder, exist_ok=True)
ids_to_remove=[]
for ind,row in tqdm(anime_to_consider.iterrows(),total=anime_to_consider.shape[0]):
    filename=f"{images_folder}/anime_{ind}_image.jpg"
    if os.path.isfile(filename):
            continue
    response = requests.get(row["main_pic"])
    if response.status_code !=200:
        ids_to_remove.append(ind)
        continue
    with Image.open(io.BytesIO(response.content)) as img:
        w,h = img.size
        s=max(w,h)
        img1 = Image.new(img.mode,(s,s),(0,0,0))
        img1.paste(img,((s-w)//2,(s-h)//2))
        img1.save(filename)
```

```python
print(ids_to_remove)
(anime_to_consider.loc[ids_to_remove]).to_csv("no_images.csv")
```

```python
anime_to_consider.drop(ids_to_remove,inplace=True)
anime_to_consider.to_csv(f"AnimeWithImages.csv")
```
