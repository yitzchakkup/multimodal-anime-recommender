```python
# Quick Recommendation System Results Plot
# =====================================

import matplotlib.pyplot as plt
import numpy as np

# Data
methods = ['Collaborative\nFiltering', 'Raw Average\nScores', 'Normalized\nAverage Scores']
cosine_scores = [0.8, 0.31, 0.15]
colors = ['#2E86AB', '#A23B72', '#F18F01']

# Create bar plot
plt.figure(figsize=(10, 6))
bars = plt.bar(methods, cosine_scores, color=colors, alpha=0.8, edgecolor='black', linewidth=1)

# Add value labels on bars
for bar, score in zip(bars, cosine_scores):
    plt.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 0.02,
             f'{score:.2f}', ha='center', va='bottom', fontweight='bold', fontsize=12)

# Formatting
plt.title('Recommendation System Performance\n(Cosine Similarity)', fontsize=16, fontweight='bold', pad=20)
plt.ylabel('Cosine Similarity Score', fontsize=12)
plt.ylim(0, 1.0)
plt.grid(axis='y', alpha=0.3)

# Add horizontal line at 0.5 for reference
plt.axhline(y=0.5, color='gray', linestyle='--', alpha=0.5, label='0.5 baseline')

plt.tight_layout()
plt.show()

# Print results
print("Recommendation System Results:")
print("-" * 30)
for method, score in zip(methods, cosine_scores):
    print(f"{method.replace(chr(10), ' ')}: {score:.2f}")
print(f"\nImprovement over best baseline: {(0.8 - 0.31) / 0.31 * 100:.1f}%")
```
